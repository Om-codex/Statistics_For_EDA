```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
delivery = pd.read_csv('deliveries.csv')
```

```python
runs = delivery.groupby('batsman').sum()['batsman_runs']
```

```python
sns.distplot(runs,color='maroon')
```

# n=50/461, batsman taken per sample
# m=1000, number of times sampling done

```python
m = 1000
n = 50
new = []
for i in range(m):
    x = np.random.choice(runs,size = n)
    new.append(x.mean())
```

```python
len(new)
```

```python
new = np.array(new)
```

```python
sns.distplot(new,color='gray')
```
